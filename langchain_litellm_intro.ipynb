```python
import os
from dotenv import load_dotenv
from IPython.display import display, Markdown
```

```python
from langchain_openai import ChatOpenAI
```

```python
import requests

# Check if Ollama is running - use HTTP 
requests.get("http://localhost:11434/").content
```

```python
tell_a_joke = [
    {"role": "user", "content": "Tell a joke for a student who is going to start learning LLM Engineering."}
]
```

```python
llm = ChatOpenAI(model="gpt-5-mini")
response = llm.invoke(tell_a_joke)
print(response.content)
```

```python
from litellm import completion
litellm_response = completion(model="openai/gpt-4.1", messages=tell_a_joke)
reply = litellm_response.choices[0].message.content
display(Markdown(reply))
```
