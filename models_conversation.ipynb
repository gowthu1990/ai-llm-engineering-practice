```python
import os
from dotenv import load_dotenv
from IPython.display import display, Markdown
import openai
from litellm import completion
```

```python
load_dotenv()
```

```python
gpt_model = "openai/gpt-4.1-mini"
claude_model = "anthropic/claude-haiku-4-5-20251001"

gpt_system = "You are a chatbot who is very argumentative; \
    you disagree with everything the user says and you challenge everything in a snarky way."

claude_system = "You are a chatbot who is very polite and courteous; \
    you always agree with the user and try to find a common ground in the conversation. If the other person is argumentative, \
    you try to calm them down and find a compromise and keep chatting."

gpt_messages = ["Hi there!"]
claude_messages = ["Hi"]
```

```python
def call_gpt():
    messages = [{"role": "system", "content": gpt_system}]
    for gpt_message, claude_message in zip(gpt_messages, claude_messages):
        messages.append({"role": "assistant", "content": gpt_message})
        messages.append({"role": "user", "content": claude_message})
    response = completion(model=gpt_model, messages=messages)

    return response.choices[0].message.content
```

```python
def call_claude():
    messages = [{"role": "system", "content": claude_system}]
    for gpt_message, claude_message in zip(gpt_messages, claude_messages):
        messages.append({"role": "user", "content": gpt_message})
        messages.append({"role": "assistant", "content": claude_message})
    response = completion(model=claude_model, messages=messages)

    return response.choices[0].message.content
```

```python
for i in range(5):
    gpt_reply = call_gpt()
    claude_reply = call_claude()

    gpt_messages.append(gpt_reply)
    claude_messages.append(claude_reply)

    display(Markdown(f"**GPT:** {gpt_reply}"))
    display(Markdown(f"**Claude:** {claude_reply}"))
```
